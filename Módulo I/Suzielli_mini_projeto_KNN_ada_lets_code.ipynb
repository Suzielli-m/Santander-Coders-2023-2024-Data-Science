**Definindo as funções**

```python
# popularidade_perfil : é um dicionário que define o nível de popularidade 
# dos perfis de investimento para possíveis empates de distância entre 
# dois ou mais vizinhos
popularidade_perfil = {
    'Conservador': 3,
    'Moderado': 2,
    'Agressivo': 1  # Exemplo: Agressivo é o menos popular
}

# distância_euclidiana : uma função para calcular a distância Euclidiana
# entre dois pontos
def distancia_euclidiana(ponto1, ponto2):
    soma_quadrados = 0
    for i in range(len(ponto1)):  # Considera todas as coordenadas
        soma_quadrados += (ponto1[i] - ponto2[i]) ** 2
    return soma_quadrados ** 0.5

# comparar_distancia : retorna o segundo elemento de uma tupla. É utilizada 
# para ordenar uma lista de tuplas com base no segundo elemento (distância para
# o presente problema)
def comparar_distancia(item):
    return item[1]

# moda : calcula o valor mais frequente de uma lista
def moda(lista):
    frequencia = {}
    for item in lista:
        if item in frequencia:
            frequencia[item] += 1
        else:
            frequencia[item] = 1

    maior_frequencia = 0
    moda = None
    for chave, valor in frequencia.items():
        if valor > maior_frequencia:
            maior_frequencia = valor
            moda = chave
    return moda

# seleciona_popularidade: retorna a popularidade de um perfil com base no 
# dicionário "popularidade_perfil"
def seleciona_popularidade(perfil):
    return popularidade_perfil[perfil]

# escolher_perfil_mais_provavel : Ordena uma lista de perfis com base em sua 
# popularidade (mais popular -> menos popilar) e retorna o primeiro perfil da 
# lista ordenada. É usada para desempatar quando há múltiplos perfis com a mesma 
# menor distância
def escolher_perfil_mais_provavel(perfis_empatados):
    # Ordena os perfis empatados pela popularidade
    perfis_empatados.sort(key=seleciona_popularidade, reverse=True)
    # Escolhe o primeiro perfil da lista ordenada
    perfil_escolhido = perfis_empatados[0]
    return perfil_escolhido

def knn(data, ponto_referencia, k):
    distancias = []

    # Calcula as distâncias
    for ponto in data:
        distancia = distancia_euclidiana(ponto[2], ponto_referencia)
        distancias.append((ponto, distancia))

    # Ordena as distâncias
    distancias.sort(key=comparar_distancia)
    vizinhos = distancias[:k]

    # Verifica se há empate nas menores distâncias
    menor_distancia = vizinhos[0][1]
    perfis_empatados = []
    for vizinho in vizinhos:
        if vizinho[1] == menor_distancia:
            perfis_empatados.append(vizinho[0][1])

    if len(perfis_empatados) > 1:
        return escolher_perfil_mais_provavel(perfis_empatados)
    else:
        # Calcula a moda das classes dos vizinhos
        classes = []
        for vizinho in vizinhos:
            classes.append(vizinho[0][1])
        return moda(classes)

```
